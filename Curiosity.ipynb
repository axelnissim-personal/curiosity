🧠 Curiosity-Driven Knowledge Exploration System
Overview
This project is a curiosity-driven knowledge exploration system named Curiosity, inspired by the concepts outlined in the paper "Curiosity Is All You Need: A Comprehensive Framework for Intrinsically Motivated Language Models".
The system autonomously explores concepts and relationships, building a dynamic knowledge graph. It uses a combination of probabilistic reasoning, reinforcement learning, and natural language processing to generate hypotheses, refine its understanding, and produce a reflective narrative of its intellectual journey. The output is presented in an HTML report, which includes a visual representation of the knowledge graph, a list of hypotheses, and a creative essay.
The system is designed to be highly observable, addressing the opacity of traditional Large Language Models (LLMs) by providing transparency into its reasoning process. Each hypothesis is annotated with confidence scores, frequency metrics, and relationship classifications, making the system's outputs interpretable and trustworthy.
Features
🔍 Curiosity-Driven Exploration

The system autonomously explores concepts and relationships, driven by curiosity scores that balance novelty, uncertainty, and reward
It generates hypotheses about relationships between concepts and validates them using an LLM

🕸️ Dynamic Knowledge Graph

The system builds and updates a knowledge graph that visually represents the relationships between concepts
Each node (concept) and edge (relationship) is annotated with metadata such as confidence scores, frequency, and classification types

🧪 Hypothesis Tracking

The system tracks and numbers each hypothesis, providing a clear audit trail of its reasoning process
Hypotheses are classified into types (e.g., Supertype-Subtype, Causal, Functional, etc.), adding semantic context to the knowledge graph

📝 Reflective Narrative

The system generates a creative essay that reflects on its intellectual journey, tying together hypotheses and insights
This narrative provides a human-readable summary of the system's exploration

📊 HTML Report
The system produces an HTML report that includes:

A visual representation of the knowledge graph
A list of seed concepts and their relationships
A detailed summary of hypotheses, including confidence scores and classifications
A creative essay that narrates the system's exploration

👁️ Observability

The system is highly observable, providing transparency into its reasoning process
This makes the system's outputs interpretable and trustworthy, addressing the opacity of traditional LLMs

Prototype Nature and Constraints
Prototype Showcasing a Concept
This system is a prototype designed to showcase the power of curiosity-driven learning and dynamic knowledge graph construction. It demonstrates how AI can autonomously explore and reason about abstract concepts, generate hypotheses, and reflect on its own learning process.
Artificial Constraint: 45 Hypotheses
To make the prototype manageable and focused, the system is artificially constrained to generate a maximum of 45 hypotheses. This constraint ensures that the exploration process is concise and that the output can be easily analyzed and interpreted.
Infinite Learning Potential
Unlike traditional LLMs, which are constrained to a fixed dataset and a specific point in time, this system has the potential to learn infinitely and in a greedy manner. It can continuously update its knowledge graph in real time, incorporating new information and refining its understanding as it explores.
Real-Time Learning
The system is designed to learn in real time, updating its knowledge graph on the fly as new hypotheses are generated and validated. This makes it a powerful tool for dynamic and evolving domains, where new information is constantly being discovered.

```python
!pip install ratelimit
import networkx as nx
import torch
import random
import re
import time
import hashlib
import google.generativeai as genai
from sentence_transformers import SentenceTransformer
from ratelimit import limits, sleep_and_retry
from google.colab import userdata  # Import the userdata module
from IPython.display import Image  # For inserting the curiosity cat
import datetime
import io
import contextlib
import matplotlib.pyplot as plt
import base64  # For embedding the image

# ANSI color codes
BLUE = '\033[94m'
RED = '\033[91m'
RESET = '\033[0m'

# HTML color mapping
ANSI_TO_HTML = {
    BLUE: 'blue',
    RED: 'red',
    RESET: 'black'  # Default color
}

# Store output lines for HTML
html_output_lines = []

def print_curiosity(message):
    global html_output_lines
    colored_message = f"{BLUE}Curiosity: {message}{RESET}"
    print(colored_message)
    html_output_lines.append(colored_message)

def print_llm_output(message, model_name="LLM"):
    global html_output_lines
    colored_message = f"{RED}LLM Output ({model_name}): {message}{RESET}"
    print(colored_message)
    html_output_lines.append(colored_message)

def ansi_to_html(text_lines):
    html_formatted_lines = []
    for line in text_lines:
        for ansi, html_color in ANSI_TO_HTML.items():
            line = line.replace(ansi, f'<span style="color:{html_color};">')
        line = line.replace(RESET, '</span>')
        html_formatted_lines.append(line)
    return "<br>".join(html_formatted_lines) + '</span>'

class CuriosityModule:
    def __init__(self, intelligence):
        self.intel = intelligence
        self.learning_priority = {}
        self.hypothesis_counter = 0  # Initialize a counter for hypotheses

    def calculate_curiosity(self, concept):
        try:
            uncertainty = 1 - self._get_confidence(concept)
            reward_grad = abs(self.intel.rewards.get(concept, 0.5) - 0.5)
            novelty = self._calculate_novelty(concept)
            age_penalty = 0.05 * len(list(nx.descendants(self.intel.knowledge_graph, concept)))
            return (uncertainty * 0.3) + (reward_grad * 0.2) + (novelty * 0.5) - age_penalty
        except Exception as e:
            print_curiosity(f"Calculation error: {str(e)}")
            return 0.5

    def _get_confidence(self, concept):
        response = self.intel.probabilistic_reasoning(
            concept,
            "your internal understanding and the information in the knowledge graph"
        )
        return max(0.0, min(1.0, response))

    def _calculate_novelty(self, concept):
        try:
            if concept not in self.intel.embeddings:
                return 1.0

            base_embed = self.intel.embeddings[concept]
            similarities = []
            for other, embed in self.intel.embeddings.items():
                if other != concept:
                    sim = torch.cosine_similarity(base_embed, embed, dim=0)
                    similarities.append(sim.item())

            return 1 - max(similarities) if similarities else 1.0
        except:
            return 0.5

    def explore(self, top_n=2):
        try:
            self.learning_priority = {
                c: self.calculate_curiosity(c)
                for c in self.intel.knowledge_graph.nodes
            }

            print_curiosity("\nCuriosity Priorities:")
            for concept, score in sorted(self.learning_priority.items(),
                                                key=lambda x: x[1], reverse=True)[:5]:
                print_curiosity(f"- {concept}: {score:.2f}")

            targets = sorted(self.learning_priority,
                             key=self.learning_priority.get,
                             reverse=True)[:top_n]

            for target in targets:
                print_curiosity(f"\nExploring from: {target}")
                self.targeted_knowledge_expansion(target)  # Corrected line

            return True
        except Exception as e:
            print_curiosity(f"Exploration failed: {str(e)}")
            return False

    def targeted_knowledge_expansion(self, base_concept):
        print_curiosity(f"\n=== Expanding from {base_concept} ===")
        prompts = [
            f"""Suggest 3 specific concepts related to '{base_concept}' using EXACT format:
            Concept: [Name], Relation: [Relationship]
            Example:
            Concept: Black Holes, Relation: extreme manifestation of
            Only return concepts in this format, no other text!""",

            f"""Generate novel associations for '{base_concept}' using format:
            Concept: [Name], Relation: [Relationship]
            Important: Use only the exact format with no additional explanations!""",

            f"""Propose connected concepts for '{base_concept}' using:
            Concept: [Name], Relation: [Relationship]
            Return only concept-relation pairs, no other text!"""
        ]

        prompt = random.choice(prompts)
        print_curiosity(f"Expansion Prompt: {prompt}")  # Added print statement
        response = self.intel.query_llm(prompt)  # Updated call
        if not response:
            return False

        parsed = re.findall(
            r'(?i)Concept:\s*([^\n\r]+?)\s*,\s*Relation:\s*([^\n\r]+?)(?=\n|Concept:|$)',
            response
        )

        added_count = 0
        for concept, relation in parsed[:3]:
            clean_concept = self.clean_concept(concept)
            clean_relation = self.clean_relation(relation)

            if (len(clean_concept) >= 3 and
                    len(clean_relation) >= 3 and
                    self.intel.validate_relation(clean_relation)):  # Corrected line
                print_curiosity(f"Parsed: {clean_concept} → {clean_relation}")
                if self.intel.add_concept(clean_concept, {clean_relation: base_concept}):
                    self._update_learning_rewards(base_concept, clean_concept)
                    added_count += 1

        return added_count > 0

    def explore_internal_links(self, top_n=5):
        nodes = list(self.intel.knowledge_graph.nodes)
        if len(nodes) < 2:
            return False

        # Get top N curiosity nodes to focus on
        sorted_curiosity = sorted(self.learning_priority.items(), key=lambda item: item[1], reverse=True)
        focused_nodes = [item[0] for item in sorted_curiosity[:top_n]]

        explored_pairs = set()
        added_link = False

        for node_a_focus in focused_nodes:
            for node_b in nodes:
                if node_a_focus == node_b:
                    continue

                # Ensure we don't explore the same pair twice (order doesn't matter for the check)
                pair = tuple(sorted((node_a_focus, node_b)))
                if pair in explored_pairs:
                    continue
                explored_pairs.add(pair)

                # --- First Prompt: Get the Relationship ---
                prompt_relation = f"""What is a possible relationship between '{node_a_focus}' and '{node_b}'? The relationship should be strictly directional from '{node_a_focus}' to '{node_b}'. Answer by providing a concise phrase describing the relationship from '{node_a_focus}' to '{node_b}', starting with a verb (e.g., 'is a type of', 'causes', 'is related to'). Make sure you don't repeat neither '{node_a_focus}' nor '{node_b}' in your description. IMPORTANT: If there is no direct or significant relationship, state 'No significant relationship'. Never provide multi-sentence answers. If you have many possibilities for answering only state the answer that feels more correct to you, but don't suggest multiple options. Do not finalize your answer with a period. NEVER include the word IT in your answer."""
                print_curiosity(f"Exploring link (Relation): {prompt_relation}")
                response_relation = self.intel.query_llm(prompt_relation)

                if response_relation and "No significant relationship" not in response_relation:
                    relation_phrase = self.clean_relation(response_relation.strip())

                    if self.intel.validate_relation(relation_phrase):
                        # --- Second Prompt: Get the Confidence Score ---
                        prompt_confidence = f"""On a scale of 0 to 1, how certain are you about the stated fact '{node_a_focus}' '{relation_phrase}' '{node_b}' being true? Please provide only a numerical value between 0 and 1 with no further explanation."""
                        print_curiosity(f"Exploring link (Confidence): {prompt_confidence}")
                        response_confidence = self.intel.query_llm(prompt_confidence)

                        # --- Third Prompt: Classify the Relationship ---
                        prompt_classification = f"""Classify the relationship '{node_a_focus}' '{relation_phrase}' '{node_b}' using one of the following types: Supertype-Subtype (Is-a), Part-Whole (Has-a / Part-of), Containment (Includes / Belongs-to), Causal (Cause-Effect), Functional (Enables / Prevents), Dependency (Needs / Requires), Usage (Used-for), Influence (Affects / Shapes), Synonymy (Same-as / Equivalent-to), Antonymy (Opposite-of), Analogy (Similar-to / Like), Degree (More/Less than), Sequence (Before-After), Simultaneity (Occurs-with / Coexists-with), Location (Located-in / Found-at), Direction (Moves-toward / Away-from), Equivalence (≡ / If-and-only-if), Implication (→ / If-then), Exclusion (Mutually Exclusive), Set Membership (Element-of / Subset-of), Conceptual Composition (Made-of / Constructed-from), Abstraction (Abstracts-from / Represents), Perspective (Viewed-as), or UNDEFINED. Provide only the type, no additional text."""
                        print_curiosity(f"Exploring link (Classification): {prompt_classification}")
                        response_classification = self.intel.query_llm(prompt_classification)

                        confidence = None
                        if response_confidence:
                            numbers = re.findall(r'\d?\.?\d+', response_confidence)
                            if numbers:
                                try:
                                    confidence = float(numbers[0])
                                    confidence = max(0.0, min(1.0, confidence))  # Ensure it's within 0 and 1
                                except ValueError:
                                    print_curiosity(f"Could not parse confidence score: {response_confidence}")

                        edge_exists = False
                        existing_edge_data = None
                        for u, v, data in self.intel.knowledge_graph.edges(data=True):
                            if u == node_a_focus and v == node_b and data.get('relation') == relation_phrase:
                                edge_exists = True
                                existing_edge_data = data
                                break

                        if edge_exists:
                            if 'frequency' in existing_edge_data:
                                existing_edge_data['frequency'] += 1
                            else:
                                existing_edge_data['frequency'] = 2  # First repeat
                            print_curiosity(f"Frequency of link {node_a_focus} → {relation_phrase} → {node_b} increased to {existing_edge_data['frequency']}")
                            added_link = True  # Consider this an update
                        else:
                            self.hypothesis_counter += 1
                            edge_data = {
                                'relation': relation_phrase,
                                'hypothesis': f"Relationship: {response_relation.strip()}",  # Store the relation response
                                'frequency': 1,
                                'classification': response_classification.strip() if response_classification else 'UNDEFINED'
                            }
                            if confidence is not None:
                                edge_data['confidence'] = confidence
                                edge_data['hypothesis'] += f", Confidence: {response_confidence.strip()}"  # Append confidence response
                            self.intel.knowledge_graph.add_edge(
                                node_a_focus,
                                node_b,
                                **edge_data
                            )
                            print_curiosity(f"Hypothesis created: {self.hypothesis_counter}")
                            print_curiosity(f"Added internal link: {node_a_focus} → {relation_phrase} → {node_b} (Confidence: {confidence if confidence is not None else 'N/A'}, Classification: {edge_data['classification']})")
                            added_link = True
                            self._update_learning_rewards(node_a_focus, node_b)
                            self._update_learning_rewards(node_b, node_a_focus)

                            # Stop after 45 hypotheses
                            if self.hypothesis_counter >= 45:
                                return added_link
        return added_link

    def clean_concept(self, text):
        return re.sub(r'[\*\"\']', '', text).strip().title()

    def clean_relation(self, text):
        return re.sub(r'[\*\"\'\(\)]', '', text).strip().lower()

    def _update_learning_rewards(self, base_concept, new_concept):
        self.intel.reinforcement_learning(base_concept, min(1.0, self.intel.rewards[base_concept] + 0.15))
        self.intel.reinforcement_learning(new_concept, 0.7)

class UniversalIntelligence:
    def __init__(self):
        self.knowledge_graph = nx.DiGraph()
        self.embeddings = {}
        self.rewards = {}
        self.embedding_model = SentenceTransformer('all-MiniLM-L6-v2')
        self.generation_config = {
            "temperature": 0.9,
            "top_p": 0.95,
            "top_k": 50,
            "max_output_tokens": 1024
        }
        self.description_cache = {}
        self.learning_interval = 10
        self.last_learned = time.time()
        self.curiosity = CuriosityModule(self)
        self.explored_relations = set()
        self.model_name = 'gemini-2.0-flash'
        self.initial_concepts_data = {}  # Store initial concepts and relations
        self.seed_start_time = 0
        self.seed_end_time = 0
        self.llm_wait_time = 0
        self.rate_limit_wait_time = 0

    def get_cache_key(self, concept):
        return hashlib.md5(concept.encode()).hexdigest()

    def seed_initial_concepts(self, initial_concepts):  # Modify to accept initial_concepts
        print_curiosity("\n--- Seeding Initial Concepts ---")
        self.seed_start_time = time.time()
        for concept, relations in initial_concepts:
            self.initial_concepts_data[concept] = relations
            if self.add_concept(concept, relations):
                print_curiosity(f"✓ {concept}")
            else:
                print_curiosity(f"✗ {concept}")
        self.seed_end_time = time.time()

    @sleep_and_retry
    @limits(calls=15, period=60)
    def query_llm(self, prompt):
        if GENAI_API_KEY is None:
            print_curiosity("LLM API key not configured. Skipping query.")
            return None
        print_curiosity(f"Querying: {prompt}")
        try:
            start_wait = time.time()
            model = genai.GenerativeModel(self.model_name,
                                            generation_config=self.generation_config)
            wait_time = time.time() - start_wait
            self.llm_wait_time += wait_time
            print_curiosity("Waiting for rate limiter...")
            response = model.generate_content(prompt)
            if response.text:
                print_llm_output(response.text, model_name=self.model_name)
                return response.text
            return None
        except Exception as e:
            if "rate limit" in str(e).lower() or "quota" in str(e).lower():
                print_curiosity(f"Rate limit reached. Error: {str(e)}")
                wait_start = time.time()
                time.sleep(60)
                wait_end = time.time()
                self.rate_limit_wait_time += (wait_end - wait_start)
                raise
            else:
                print(f"API error: {str(e)}")
                return None

    def add_concept(self, concept, relations=None):
        try:
            concept = concept.strip().title()
            if not re.match(r"^[\w\s'\-\(\)]{3,50}$", concept):
                print_curiosity(f"Invalid concept format: {concept}")
                return False

            cache_key = self.get_cache_key(concept)
            if cache_key not in self.description_cache:
                description = self.query_llm(f"Describe the concept of {concept} in one sentence. Make sure you're describing a concept. Don't provide anything else.")  # Updated call
                if not description:
                    return False
                self.description_cache[cache_key] = description

            embedding = self.embedding_model.encode(self.description_cache[cache_key])
            self.embeddings[concept] = torch.tensor(embedding)
            self.rewards[concept] = 0.5
            self.knowledge_graph.add_node(concept)
            print_curiosity(f"Added concept: {concept}")

            if relations:
                for rel, target in relations.items():
                    target = target.strip().title()
                    if not self.validate_relation(rel) or not target:
                        continue

                    if target not in self.embeddings:
                        self.add_concept(target)

                    if nx.has_path(self.knowledge_graph, target, concept):
                        print_curiosity(f"Prevented circular relation: {target} → {concept}")
                        continue

                    relation_tuple = (concept, rel, target)
                    if relation_tuple not in self.explored_relations:
                        self.knowledge_graph.add_edge(concept, target, relation=rel)
                        self.explored_relations.add(relation_tuple)
                        print_curiosity(f"Added relation: {concept} → {rel} → {target}")
            return True
        except Exception as e:
            print_curiosity(f"Error adding {concept}: {str(e)}")
            return False

    def validate_relation(self, relation):
        relation = relation.lower().strip()
        banned_terms = {'associated', 'linked', 'connection'}
        return (
            re.match(r"^[\w\s-]{3,30}$", relation) and
            not any(term in relation for term in banned_terms) and
            len(relation.split()) <= 4
        )

    def probabilistic_reasoning(self, concept, reasoning_context):
        prompt = f"On a scale of 0 to 1, based on '{reasoning_context}', how certain are you about the concept of '{concept}'? Please provide only the numerical value."
        print_curiosity(f"Probability Prompt: {prompt}")
        response = self.query_llm(prompt)
        numbers = re.findall(r'\d?\.?\d+', response or "")
        if numbers:
            try:
                value = float(numbers[0])
                return max(0.0, min(1.0, value))
            except ValueError:
                pass
        return 0.5

    def reinforcement_learning(self, concept, reward):
        if concept in self.rewards:
            self.rewards[concept] = 0.85 * self.rewards[concept] + 0.15 * reward
            self.rewards[concept] = max(0.1, min(0.9, self.rewards[concept]))

    def show_graph_stats(self):
        print_curiosity("\nKnowledge Graph Summary:")
        print_curiosity(f"Nodes: {len(self.knowledge_graph.nodes)}")
        print_curiosity(f"Edges: {len(self.knowledge_graph.edges)}")
        print_curiosity("Top Concepts by Reward:")
        for concept in sorted(self.rewards, key=self.rewards.get, reverse=True)[:5]:
            print_curiosity(f"- {concept}: {self.rewards[concept]:.2f}")

    def generate_knowledge_graph_summary(self):
        """
        Generates a textual description of the knowledge graph, including explored nodes,
        relationships, confidence levels, and whether they were promoted to the final worldview.
        """
        summary = []

        # Header
        summary.append("=== Knowledge Graph Summary ===")
        summary.append("\n--- Explored Nodes and Relationships ---")

        # Track explored nodes and relationships
        explored_nodes = set()
        explored_relationships = []

        # Iterate through all edges in the graph
        for u, v, data in self.knowledge_graph.edges(data=True):
            explored_nodes.add(u)
            explored_nodes.add(v)
            relation = data.get('relation', 'Unknown')
            confidence = data.get('confidence', 'N/A')
            frequency = data.get('frequency', 0)
            is_hypothesis = 'hypothesis' in data

            # Add relationship details
            relationship_description = (
                f"- Relationship: {u} → {relation} → {v}\n"
                f"  Confidence: {confidence}, Frequency: {frequency}\n"
                f"  Status: {'Promoted to Worldview' if is_hypothesis else 'Not Promoted'}"
            )
            explored_relationships.append(relationship_description)

        # Add explored nodes section
        summary.append("\nExplored Nodes:")
        for node in explored_nodes:
            reward = self.rewards.get(node, 'N/A')
            summary.append(f"- {node} (Reward: {reward:.2f})")

        # Add explored relationships section
        summary.append("\nExplored Relationships:")
        summary.extend(explored_relationships)

        # Add final worldview hypotheses section
        summary.append("\n--- Final Worldview Hypotheses ---")
        worldview_hypotheses = []
        for u, v, data in self.knowledge_graph.edges(data=True):
            if 'hypothesis' in data:
                hypothesis_text = data['hypothesis']
                relation = data.get('relation', 'No Relation')
                confidence = data.get('confidence', 'N/A')
                frequency = data.get('frequency', 0)
                worldview_hypotheses.append(
                    f"- Hypothesis: {u} → {relation} → {v}\n"
                    f"  Confidence: {confidence}, Frequency: {frequency}\n"
                    f"  Details: {hypothesis_text}"
                )

        if worldview_hypotheses:
            summary.extend(worldview_hypotheses)
        else:
            summary.append("No hypotheses were promoted to the final worldview.")

        return "\n".join(summary)

    def autonomous_learning_cycle(self):
        if GENAI_API_KEY is None:
            return  # Skip learning cycle if API key is not available
        if time.time() - self.last_learned >= self.learning_interval:
            print_curiosity("\n=== Curiosity-Driven Learning Cycle ===")
            if self.curiosity.explore(top_n=3):
                pass  # Exploration of new concepts happened
            if self.curiosity.explore_internal_links(top_n=5):  # Explore links between existing concepts
                pass
            self.show_graph_stats()
            self.last_learned = time.time()

    def print_all_hypotheses(self):
        print_curiosity("\n--- Worldview Hypotheses ---")
        for u, v, data in self.knowledge_graph.edges(data=True):
            if 'hypothesis' in data:
                hypothesis_text = data['hypothesis']
                relation = data.get('relation', 'No Relation')
                confidence = data.get('confidence', 'N/A')
                frequency = data.get('frequency', 0)
                print_curiosity(f"Hypothesis: {u} → {relation} → {v} (Confidence: {confidence}, Frequency: {frequency})")
                print_llm_output(f"  Text: {hypothesis_text}")

    def print_time_utilization(self):
        seed_duration = self.seed_end_time - self.seed_start_time
        num_seed_concepts = len(self.initial_concepts_data)
        num_seed_relations = sum(len(rels) for rels in self.initial_concepts_data.values())

        print_curiosity("\n--- Time Utilization Metrics ---")
        print_curiosity(f"Seed Ingestion:")
        print_curiosity(f"- Concepts: {num_seed_concepts}")
        print_curiosity(f"- Relationships: {num_seed_relations}")
        print_curiosity(f"- Time Taken: {seed_duration:.2f} seconds")
        print_curiosity(f"LLM Wait Time: {self.llm_wait_time:.2f} seconds")
        print_curiosity(f"Rate Limit Wait Time: {self.rate_limit_wait_time:.2f} seconds")

if __name__ == "__main__":
    # Replace with your actual API key
    GENAI_API_KEY = userdata.get('GOOGLE_API_KEY')
    genai.configure(api_key=GENAI_API_KEY)

    ui = UniversalIntelligence()
    initial_concepts = [
        ("Consciousness", {"involves": "Subjective Experience"}),
        ("Neural Activity", {"correlates with": "Conscious States"}),
        ("Curiosity", {"involves": "Intrinsic Motivation"}),
        ("Absolute certainty", {"is contrary to": "Curiosity"})
    ]

    initial_concepts_list = [item[0] for item in initial_concepts]  # Store initial concepts

    interrupted = False  # Flag to track if the script was interrupted

    ui.seed_initial_concepts(initial_concepts)  # Call the modified seeding method

    print_curiosity("\n--- Starting Autonomous Learning ---")
    try:
        while True:
            try:
                ui.autonomous_learning_cycle()
                time.sleep(10)
                if random.random() < 0.25:
                    ui.show_graph_stats()
                if ui.curiosity.hypothesis_counter >= 45:
                    print_curiosity("\n--- Reached 45 Hypotheses ---")
                    break
            except KeyboardInterrupt:
                interrupted = True
                print_curiosity("\n--- Final Knowledge Summary (Interrupted) ---")
                ui.show_graph_stats()
                break  # Exit the outer loop as well
    finally:
        # Generate HTML content
        timestamp = datetime.datetime.now().strftime("%d%m%Y%H%M%S")
        filename = f"Curiosity{timestamp}.html"
        color_coded_output = ansi_to_html(html_output_lines)
        html_content = f"""
        <!DOCTYPE html>
        <html>
        <head>
            <title>Curiosity Run - {timestamp}</title>
            <style>
                .center {{
                    display: flex;
                    justify-content: center;
                    margin: 20px 0;
                }}
                .section-explanation {{
                    background-color: #f9f9f9;
                    padding: 15px;
                    border-left: 5px solid #007bff;
                    margin: 20px 0;
                }}
            </style>
        </head>
        <body>
            <h1>Curiosity Run - {timestamp}</h1>
            <div class="center">
                <img src="https://media.licdn.com/dms/image/v2/D4E22AQHMjJaPnz-6JA/feedshare-shrink_2048_1536/B4EZWy4eIaGYAw-/0/1742462903219?e=1745452800&v=beta&t=zUVPFoVYjiIcVLOjrKaxl4yOEW1429fi1c4OB4N70-Y" alt="Curiosity Image" style="width: 50%;">
            </div>

            <!-- Section Explanations -->
            <div class="section-explanation">
                <h2>How to Understand This Document</h2>
                <p>
                    This document is a record of an autonomous learning system's exploration of concepts, starting from a set of seed concepts and expanding through curiosity-driven hypothesis generation. Below is a breakdown of each section:
                </p>
                <ul>
                    <li><strong>Knowledge Graph Diagram</strong>: A visual representation of the system's exploration. It shows how concepts are connected, with nodes representing concepts and edges representing relationships. The graph evolves as the system explores new ideas.</li>
                    <li><strong>Seed Concepts</strong>: The initial set of concepts and relationships provided to the system. These serve as the foundation for exploration.</li>
                    <li><strong>Most Recent Worldview</strong>: A list of hypotheses that met certain requirements and were incorporated into the knowledge graph. Each hypothesis is numbered and includes details about the relationship type, confidence, and frequency.</li>
                    <li><strong>Creative Essay</strong>: A narrative generated by the system, reflecting on its intellectual journey and the insights gained from exploring the hypotheses.</li>
                </ul>
            </div>

            <pre>{color_coded_output}</pre>

            <!-- Knowledge Graph Diagram -->
            <h2>Knowledge Graph Diagram</h2>
            <div class="section-explanation">
                <p>
                    The graph below represents the system's exploration of concepts and relationships. It starts with the seed concepts (highlighted in blue) and expands to include new concepts (highlighted in green) and hypotheses (edges with labels). The graph is a snapshot of the system's evolving understanding.
                </p>
            </div>
            <img src="data:image/png;base64," alt="Knowledge Graph">

            <!-- Seed Concepts -->
            <h2>Seed Concepts</h2>
            <div class="section-explanation">
                <p>
                    These are the initial concepts and relationships provided to the system. They serve as the starting point for exploration and are highlighted in blue in the graph.
                </p>
            </div>
            <ul>
        """

        # Add Seed Concepts Section
        for concept, relations in ui.initial_concepts_data.items():
            html_content += f"<li><strong>{concept}</strong>"
            if relations:
                html_content += "<ul>"
                for relation, target in relations.items():
                    html_content += f"<li>→ {relation} → {target}</li>"
                html_content += "</ul>"
            html_content += "</li>"
        html_content += "</ul>"

        # Add Most Recent Worldview Section
        html_content += """
            <h2 id='most-recent-worldview'>Most Recent Worldview</h2>
            <div class="section-explanation">
                <p>
                    This section lists the hypotheses that were incorporated into the knowledge graph. Each hypothesis is numbered and includes details about the relationship type, confidence, and frequency. These hypotheses represent the system's current understanding of the explored concepts.
                </p>
            </div>
            <ol>
        """
        for idx, (u, v, data) in enumerate(ui.knowledge_graph.edges(data=True), start=1):
            if 'hypothesis' in data:
                hypothesis_text = data['hypothesis'].replace('\n', '<br>')
                relation = data.get('relation', 'No Relation')
                confidence = data.get('confidence', 'N/A')
                frequency = data.get('frequency', 0)
                classification = data.get('classification', 'UNDEFINED')
                html_content += f"""
                <li>
                    <strong>{u} → {relation} → {v}</strong><br>
                    Classification: {classification}<br>
                    Confidence: {confidence}, Frequency: {frequency}<br>
                    Details: {hypothesis_text}
                </li>
                """
        html_content += "</ol>"

        # Generate knowledge graph diagram
        plt.figure(figsize=(12, 12))
        pos = nx.spring_layout(ui.knowledge_graph)
        node_colors = ['lightblue' if node in initial_concepts_list else 'lightgreen' for node in ui.knowledge_graph.nodes]
        nx.draw(ui.knowledge_graph, pos, with_labels=True, node_color=node_colors, node_size=3000, font_size=10, font_weight='bold', arrows=True)
        edge_labels = nx.get_edge_attributes(ui.knowledge_graph, 'relation')
        nx.draw_networkx_edge_labels(ui.knowledge_graph, pos, edge_labels=edge_labels)

        # Save the graph to a PNG image in memory
        buffer = io.BytesIO()
        plt.savefig(buffer, format='png')
        buffer.seek(0)
        image_base64 = base64.b64encode(buffer.read()).decode('utf-8')
        plt.close()

        # Embed the graph in the HTML content
        html_content = html_content.replace('<img src="data:image/png;base64," alt="Knowledge Graph">', f'<img src="data:image/png;base64,{image_base64}" alt="Knowledge Graph">')

        # Add Creative Essay Section
        html_content += """
            <h2>Creative Essay: The Intellectual Journey</h2>
            <div class="section-explanation">
                <p>
                    The following essay is a creative reflection generated by the system. It tells the story of how the seed concepts prompted the evaluation of hypotheses and the insights gained along the way. Each sentence is marked with the hypothesis number it references.
                </p>
            </div>
            <div style="background-color: #f0f8ff; padding: 15px; border-radius: 5px;">
        """

        # Generate the creative essay
        seed_concepts_text = "Seed Concepts:\n"
        for concept, relations in ui.initial_concepts_data.items():
            seed_concepts_text += f"- {concept}\n"
            if relations:
                for relation, target in relations.items():
                    seed_concepts_text += f"  → {relation} → {target}\n"

        worldview_text = "Most Recent Worldview:\n"
        for idx, (u, v, data) in enumerate(ui.knowledge_graph.edges(data=True), start=1):
            if 'hypothesis' in data:
                hypothesis_text = data['hypothesis'].replace('\n', ' ')
                relation = data.get('relation', 'No Relation')
                confidence = data.get('confidence', 'N/A')
                frequency = data.get('frequency', 0)
                worldview_text += f"{idx}. {u} → {relation} → {v} (Confidence: {confidence}, Frequency: {frequency})\n"
                worldview_text += f"   Details: {hypothesis_text}\n"

        essay_prompt = f"""Taking the following hypotheses as input, their type of relationships, write a creative essay telling the story about how the realization of the Seed Concepts prompted you to evaluate the hypotheses (The worldview) that you will see listed here and what insights you gained. Tell it as an intellectual journey fueled by curiosity. Make it compelling. Make sure you touch upon every hypothesis, even the ones with the lowest certainty, for each sentence you write mark it with the hypothesis number you're using:

{seed_concepts_text}

{worldview_text}"""

        essay_response = ui.query_llm(essay_prompt)

        if essay_response:
            # Replace newlines with <br> before embedding in the HTML
            essay_response_cleaned = essay_response.replace('\n', '<br>')
            html_content += f"<p>{essay_response_cleaned}</p>"
        else:
            html_content += "<p>Failed to generate the creative essay.</p>"

        html_content += "</div></body></html>"

        # Write HTML content to file
        with open(filename, "w") as f:
            f.write(html_content)

        print(f"\nOutput saved to {filename}")
```
